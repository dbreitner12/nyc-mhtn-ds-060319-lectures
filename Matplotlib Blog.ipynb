```python
import matplotlib.pyplot as plt

#When using a Jupyter Notebook
%matplotlib inline 

#When using some other application
plt.show() 

#Random Data
x = [1,2,3,4,5]
y = [x**2 for x in x] 

plot_1 = plt.figure()
plot_1.add_axes([0,0,1,1]) #Where you want your plot
plot_1.subplot(x,y)

#How to set title
plot_1.set_title('Plot 1')
#How to set X-axis
plot_1.set_xlabel('X-Axis')
#How to set Y-axis
plot_1.set_ylabel('Y-Axis')

```

```python
import matplotlib.pyplot as plt
#When using a Jupyter Notebook
%matplotlib inline 
#When using some other application
plt.show() 

#Random data
x = [1,2,3,4,5]
y = [x**2 for x in x] 

#How to create a figure
fig = plt.figure()
#Size and location of plot
plot_1 = fig.add_axes([0,0,1,1])
#Customizable features
plot_1.plot(x,y, label = 'X vs Y', color = 'blue',linewidth = 5, 
            alpha = 1,linestyle = '--', marker = 'o', 
            markersize = 10, markerfacecolor = 'yellow')
plot_1.plot(y,x, label = 'Y vs X', color = 'red',linewidth = 5,
            alpha = 1, linestyle = ':', marker = 'o', 
            markersize = 10, markerfacecolor = 'green')

#How to set a title
plot_1.set_title('Plot 1')
#How to label x-axis
plot_1.set_xlabel('X-Axis')
#How to label y-axis
plot_1.set_ylabel('Y-Axis')
#How to make a legend
plot_1.legend(loc = 0, prop = {'size': 12})                             

#How to customize tickers
from matplotlib.ticker import (MultipleLocator)  
plot_1.xaxis.set_major_locator(MultipleLocator(5))
plot_1.xaxis.set_minor_locator(MultipleLocator(1))
plot_1.yaxis.set_major_locator(MultipleLocator(5))
plot_1.yaxis.set_minor_locator(MultipleLocator(1))
plot_1.tick_params(which='both', width=2)
plot_1.tick_params(which='major', length=10)
plot_1.tick_params(which='minor', length=4)

#How to add a grid
plot_1.grid()
```

```python
#How to change color of plot, understands basic colors
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green')
```

```python
#How to change line width, default is 1, can also pass in lw
#How to change transparency of line with alpha parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green', linewidth = 10, alpha = 1)
```

```python
#How to change linestyle
#Some examples: '..', ':', '--', 'steps'
#Can also pass in ls
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green', linestyle = '--')
```

```python
#How to put markers on your plot
# Some examples: 'o', 'x', '+' 
#How to change marker size
#How to change marker color

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green', marker = 'o', markersize = 20, 
        markerfacecolor = 'blue')
```

```python
#How to set x limit and y limit to see specfic parts of your plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green')

ax.set_xlim([0,2.5])
ax.set_ylim([0,10])
```
