```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = [1,2,3,4,5]
y = [x**2 for x in x]
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) #Where you want your plot
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
#Left,Bottom,Width,Height

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) #1 row by 2 columns
plt.tight_layout() #Spaces out the plots, recommend always putting at end 

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig,axes = plt.subplots(nrows = 2, ncols =1, figsize =(8,2))
#figsize(8,2) means 8 wide by 2 height
axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
```

```python
#To save a figure
fig.savefig('my_picture.png')
```

```python
#How to add a legend (label), 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label = 'X vs Y') 
ax.plot(y,x, label = 'Y vs X')

ax.legend(loc = 0)
#loc = 0 is a location code
```

```python
#How to change color of plot, understands basic colors
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green')
```

```python
#How to change line width, default is 1, can also pass in lw
#How to change transparency of line with alpha parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green', linewidth = 10, alpha = 1)
```

```python
#How to change linestyle
#Some examples: '..', ':', '--', 'steps'
#Can also pass in ls
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green', linestyle = '--')
```

```python
#How to put markers on your plot
# Some examples: 'o', 'x', '+' 
#How to change marker size
#How to change marker color

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green', marker = 'o', markersize = 20, 
        markerfacecolor = 'blue')
```

```python
#How to set x limit and y limit to see specfic parts of your plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'green')

ax.set_xlim([0,2.5])
ax.set_ylim([0,10])
```
